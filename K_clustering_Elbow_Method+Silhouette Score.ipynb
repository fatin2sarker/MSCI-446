```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


data = pd.read_csv("updated_cleaned_recipe_dataset.csv")

# Analyzing based on nutritional information
features = data[['Calories', 'Carbohydrates', 'Proteins']]

# Scaling the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Determining the optimal number of clusters
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 11)  # Typically, start from 2 to an upper limit of your choice

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(features_scaled)
    cluster_labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)

# Plotting the Elbow Method
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Nutritional Information')

# Plotting the Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Nutritional Information')

plt.tight_layout()
plt.show()

# Analyzing based on  Health Goals 

# Encoding Health Goals as binary features
health_goals_encoded = data['Health Goals'].str.get_dummies()

# Feature scaling
scaler = StandardScaler()
health_goals_scaled = scaler.fit_transform(health_goals_encoded)

# Analyzing the optimal number of clusters
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(health_goals_scaled)
    cluster_labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(health_goals_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the Elbow Method and Silhouette Scores
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Health Goals')

plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Health Goals')

plt.tight_layout()
plt.show()

#Analyzing based on Diatery Restrictions

# Encoding Dietary Restrictions as binary features
dietary_restrictions_encoded = data['Dietary Preference'].str.get_dummies()

# Feature scaling
scaler = StandardScaler()
dietary_restrictions_scaled = scaler.fit_transform(dietary_restrictions_encoded)

# Analyzing the optimal number of clusters
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(dietary_restrictions_scaled)
    cluster_labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(dietary_restrictions_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the Elbow Method and Silhouette Scores
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Dietary Restrictions')

plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Dietary Restrictions')

plt.tight_layout()
plt.show()

# Analyzing based on Recipe Category

# Encoding Recipe Category as binary features
recipe_category_encoded = data['Recipe Category'].str.get_dummies()

# Feature scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(recipe_category_encoded)

# Analyzing the optimal number of clusters
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 11)  # Example range from 2 to 10

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(features_scaled)
    cluster_labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the Elbow Method and Silhouette Scores
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Recipe Category')

plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Recipe Category')

plt.tight_layout()
plt.show()

# Analyzing on Recipe Difficulty

# Encoding Recipe Difficulty as binary features
recipe_difficulty_encoded = data['Recipe Difficulty'].str.get_dummies()

# Feature scaling
scaler = StandardScaler()
features_scaled = scaler.fit_transform(recipe_difficulty_encoded)

# Analyzing the optimal number of clusters
inertia = []
silhouette_scores = []
range_n_clusters = range(2, 11)  # Example range from 2 to 10

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(features_scaled)
    cluster_labels = kmeans.labels_
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(features_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plotting the Elbow Method and Silhouette Scores
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.plot(range_n_clusters, inertia, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Recipe Difficulty')

plt.subplot(1, 2, 2)
plt.plot(range_n_clusters, silhouette_scores, '-o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Recipe Difficulty')

plt.tight_layout()
plt.show()
```
